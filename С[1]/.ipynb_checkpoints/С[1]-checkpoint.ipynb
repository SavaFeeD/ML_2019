# Разбиваем сложные атрибуты

<b>Можно разбить Автора на стобцы: Автор ФИО и Автор Должность

```python
def split_author(x):
    arr = np.array(x.split('('))
    return arr
    
df['Автор'] = df['Автор'].apply(split_author)
df.head()
```

```python
df['Автор ФИО'] = df['Автор'].apply(lambda x: x[0])
df.head()
```

```python
df['Автор Должность'] = df['Автор'].apply(lambda x: x[1][:-1])
df.head()
```

```python
from collections import Counter
```

```python
len(Counter(df['Автор ФИО']))
```

После того как разбили столбец на два можно удалить исходный

```python
df = df.drop('Автор', axis=1)
df.head()
```

```python
X = df.drop(['Адресат'], axis=1)
y = df['Адресат']
```

```python
all_rows = []
count = 0

for i in range(df.shape[0]):
    all_rows.append([])
    for item in X.iloc[i]:
        if pd.notna(item):
            all_rows[count].append(item)
    count += 1
```

```python
from nltk.tokenize import word_tokenize
```

```python
all_tokens = []

for row in all_rows:
    all_tokens.append(pd.Series(row).apply(word_tokenize))
```

```python
void = [[] for i in range(df.shape[0])]
df_clas = pd.DataFrame({'data': void})
for i in range(df.shape[0]):
    for item in all_tokens[i]:
        df_clas['data'].iloc[i].append(item)
    
df_clas.shape
```

```python
def clean_punkt(row):
    row_clean = []
    key_list = 0
    for list_word in row:
        row_clean.append([])
    
        for word in list_word:
            if word.isalpha():
                row_clean[key_list].append(word.lower())
            
        key_list += 1
    return row_clean

df_clas['data'] = df_clas['data'].apply(clean_punkt)
df_clas.shape
```

```python
from nltk.corpus import stopwords
```

```python
sw = set(stopwords.words('russian'))

def clean_sw(row):
    arr = []
    count = 0
    
    for list_word in row:
        arr.append([])
        
        for word in list_word:
            if not word in sw:
                arr[count].append(word)
                
        count += 1
            
    return arr

df_clas['data'] = df_clas['data'].apply(clean_sw)
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
snowball = SnowballStemmer('russian')

def clean_stem(row):
    arr = []
    count = 0
    
    for list_word in row:
        arr.append([])
        
        for word in list_word:
            arr[count].append(snowball.stem(word))
                
        count += 1
    return arr

df_clas['data'] = df_clas['data'].apply(clean_stem)
```

```python
def join_list(row):
    arr = []
    count = 0
    for list_word in row:
        
        arr.append(' '.join(list_word))
        
    return arr

df_clas['data'] = df_clas['data'].apply(join_list)
```

```python
void = ['' for i in range(df.shape[0])]
df = pd.DataFrame({'short': void, 'note': void, 'fio': void, 'pos': void})

for i in range(df_clas.shape[0]):
    size = pd.Series(df_clas['data'].iloc[i]).shape[0]
    
    if size == 2:
        df['short'].iloc[i] = df_clas['data'].iloc[i][0]
        df['note'].iloc[i] = df_clas['data'].iloc[i][1]
        df['fio'].iloc[i] = df_clas['data'].iloc[i][2]
        df['pos'].iloc[i] = df_clas['data'].iloc[i][3]
    
    elif size == 3:
        df['short'].iloc[i] = df_clas['data'].iloc[i][0]
        df['fio'].iloc[i] = df_clas['data'].iloc[i][1]
        df['pos'].iloc[i] = df_clas['data'].iloc[i][2]
```

# Формируем словарь

```python
df['data'] = void

for i in range(df.shape[0]):
    df['data'].iloc[i] = ' '.join([df['short'].iloc[i], df['note'].iloc[i], df['fio'].iloc[i], df['pos'].iloc[i]])
    
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=25000)
matrix = tfidf.fit_transform(df['data'])
matrix.shape
```

```python
vocab = tfidf.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
matrix_cv = cv.fit_transform(df['data'])
matrix_cv.shape
```

# Классифицируем

```python
X = pd.DataFrame(matrix.toarray(), columns = tfidf.get_feature_names())
X_cv = pd.DataFrame(matrix_cv.toarray(), columns = cv.get_feature_names())
```

```python
from sklearn.model_selection import train_test_split
```
